```python
import pandas as pd
# Usamos la función de pandas .read_csv() y el resultado se lo asignamos a una variable
# Como argumento de .read_csv() usamos la ruta del archivo a leer.
# Esto es una de las maneras en que podemos crear un DataFrame
df_players = pd.read_csv('./drive/MyDrive/RedesNeuronales/RegresionLineal/players_20.csv')
df_players
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Tamaño del gráfico
plt.figure(figsize=(10, 6))

# Gráfico de los datos originales
def exponential_func(x, a, b):
    return a * np.exp(b * x)

# Obtener columnas x e y
x = df_players['potential']
y = df_players['value_eur']

plt.plot(x, y, marker='o', linestyle='', color='aquamarine', label='Datos Originales')
plt.xlabel('Variable independiente (x)')
plt.ylabel('Variable dependiente (y)')
plt.legend()
plt.show()


# Realizar ajuste exponencial
popt, pcov = curve_fit(exponential_func, x, y)

# Parámetros del ajuste
a_fit, b_fit = popt

# Generar datos ajustados para plot
y_fit = exponential_func(x, a_fit, b_fit)

# Graficar resultados
plt.plot(x, y_fit, marker='o', linestyle='' ,color='darkslategray', label='Ajuste exponencial')
plt.xlabel('Variable independiente (x)')
plt.ylabel('Variable dependiente (y)')
plt.legend()
plt.show()

print(f'Parámetros del ajuste: a = {a_fit}, b = {b_fit}')
```

```python
# Importar módulo matlplotlib
from matplotlib import pyplot as plt


# Leemos los datos accediendo a la columna correspondiente
# Usamos la notación de listas nombre_de_la_lista['Nombre_Columna']

x = df_players['height_cm']
y = df_players['movement_agility']

# Graficamos el diagrama de dispersión

plt.plot(x, y, marker='o', linestyle='', color='gold')
plt.show()
```

```python
# Importamos la librería numpy

import numpy as np

# Usaremos el método round() para redondear los resultados
# Sintaxis: round(valor_a_redondear, cantidad de decimales)

# Calculamos los promedios
promedio_x = round(x.mean(), 2)
promedio_y = round(y.mean(), 2)

# Calculamos la desviación estándar
desv_std_x = round(np.std(x), 2)
desv_std_y = round(np.std(y), 2)

# Calculamos la varianza
var_x = round(np.cov(x, y)[0][0], 2)

# Calculamos la covarianza
cov_x_y = round(np.cov(x, y)[0][1], 2)

# Mostramos los resultados

print(promedio_x)
print(promedio_y)
print(desv_std_x)
print(desv_std_y)
print(var_x)
print(cov_x_y)

# Calculamos el coeficiente de Correlación R de Pearson

rPearson = round( cov_x_y/(desv_std_x * desv_std_y), 2)

print("El coeficiente R de Person es: ",rPearson)

# Calculamos el estimador Beta 1 (pendiente)

beta_1 = round((cov_x_y/var_x), 2)

# Calculamos el estimador Beta 0 (intercepto)

beta_0 = round((promedio_y-beta_1*promedio_x), 2)

print(beta_1, beta_0)
```

```python
# Crear el gráfico de con la función estimada a través de la Regresión Lineal
from matplotlib import pyplot as plt

# Tamaño del gráfico
plt.figure(figsize=(10, 6))

# Gráfico de los datos originales

x = df_players['height_cm']
y = df_players['movement_agility']

plt.plot(x, y, marker='o', linestyle='', color='gold', label='Datos Originales')

# Gráfico de Función estimada
x2 = df_players['height_cm']

y2=[]
for i,e in enumerate(x):
  value = beta_0 + beta_1 * x2[i]
  y2.append(value)

plt.plot(x2, y2, marker='s', linestyle='--', color='deeppink', label='Función Estimada')
plt.xlabel('Altua')
plt.ylabel('Agilidad de movimiento')
plt.title('Regresión Lineal')
plt.legend()
plt.show()
```

```python
# Importar la librería Pandas y Numpy la que nos permite leer archivos de datos y
# manipularlos

import pandas as pd
import numpy as np

# Desde la librería sklearn.linear_model importamos LinearRegression
from sklearn.linear_model import LinearRegression
import random

# Para poder leer los datos desde nuestro DataFrame
# tenemos que indicar la columna que contiene los datos,
# luego transformarlo en un arreglo o array de Numpy
# luego transformarlo a un array de tipo 2D usando la función reshape

x = df_players['weight_kg'].to_numpy().reshape((-1,1))
y = df_players['power_strength'].to_numpy().reshape((-1,1))

# Imprimimos los datos originales

from matplotlib import pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='', color='gold', label='Datos Originales')

# Creamos una variable llamada model con la Regresión Lineal

model = LinearRegression()

# Le indicamos los parámetros para entrenamiento
model.fit(x,y)

# Le pedimos que calcule el índice de correlación R de Pearson
R_sq = model.score(x,y)

print("El coeficiente de determinación R de Pearson es: ", round(R_sq, 2))
print("El intercepto Beta_0 es: ", model.intercept_)
print("La pendiente Beta_1 es: ", model.coef_)
print("La función es estimada es: {0} + {1}x ".format(model.intercept_, model.coef_))

# Haremos una predicción
y_pred = model.predict(x)
print("Valores predichos por el modelo: ", y_pred)

# Ahora haremos una array de números al azar usando random
# esto lo haremos para probar el modelo
x_new = np.random.randint(low=45, high=105, size=25).reshape((-1,1))
print(x_new)
y_new = model.predict(x_new)
print(y_new)
plt.plot(x_new, y_new, marker='s', linestyle='--', color='deeppink', label='Prueba del modelo')

plt.title('Regresión Lineal con SkLearn')
plt.legend()
plt.show()
```
