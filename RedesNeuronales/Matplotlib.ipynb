# Importar Librerias

```python
import matplotlib.pyplot as plt
```

# Line Plot

```python
plt.figure(figsize=(5, 3))

# pais 1
x = [2016, 2017, 2018, 2019, 2020, 2021] #años
y = [42, 43, 45, 47, 48, 50] #poblacion

# pais 2
x2 = [2016, 2017, 2018, 2019, 2020, 2021] #años
y2 = [43, 43, 44, 44, 45, 45] #poblacion

# plotting
plt.plot(x, y, marker='o', linestyle='--', color='g', label='Colombia')
plt.plot(x2, y2, marker='d', linestyle='-', color='r', label='Argentina')

# agregar nombre a los ejes y titulo al grafico
plt.xlabel('Años')
plt.ylabel('Poblacion (M)')
plt.title('Años vs Poblacion')
plt.legend(loc='lower right')

plt.yticks([41, 45, 48, 51])

# guardar figura
plt.savefig('ejemplo.png')

# mostrar plot
plt.show()
```

# Subplots

```python
# ejemplo 2 subplots
fig, ax = plt.subplots(1, 2, sharey=True, figsize=(12, 4)) #figsize modifica el tamaño del plot

ax[0].plot(x, y, color='g', label='Colombia')
ax[0].legend()

ax[1].plot(x2, y2, color='r', label='Argentina')
ax[1].legend()

plt.show()

# ------
# ejemplo +2 subplots

# fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10, 10))

# ax[0][0].plot(x, y, 'o-', c='r', label='Colombia') # fila 1 - columna 0 (columnas comienzan en 0)
# ax[0][2].plot(x2, y2, 'v--', c='b', label='Argentina') # fila 1 - columna 2

# plt.show()
```

# Bar Plots

```python
x = ['Argentina', 'Colombia', 'Peru'] # paises
y = [40, 50, 33] # poblacion (data numérica)
plt.figure(figsize=(12, 6))
plt.bar(x, y)
plt.show()
```

# Piecharts

```python
plt.pie(y, labels=x, autopct='%.0f %%') # "autopct" agrega los porcentajes a la gráfica
plt.show()
```

# Histogram

```python
edades = [15, 16, 17, 20, 21, 21, 22, 23, 24 ,25, 26, 30, 31, 32, 35]
bins = [15, 20, 25, 30, 35]
#los bins determinan los intervalos de las barras


plt.hist(edades, bins, edgecolor='black')
plt.show()
```

# Boxplots

```python
edades = [15, 16, 17, 20, 21, 21, 22, 23, 24 ,25, 26, 30, 31, 32, 35, 70] # 70 es un valor atípico en esta lista

plt.boxplot(edades)
plt.show()
```
