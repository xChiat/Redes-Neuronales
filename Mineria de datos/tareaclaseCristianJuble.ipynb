```r
# Este código genera un número aleatorio con distribución normal y utiliza una estructura 'if-else' para determinar si el número es mayor o igual a 1, o menor que 1.
valor <- rnorm(1)
print(valor)
if (valor >= 1) {
  print("Número mayor o igual a 1")
}else{
  print("Número menor a 1")
}
```

```r
# Este código genera un número aleatorio con distribución normal y utiliza una estructura 'if-else if-else' para clasificar si el número es mayor que 0, igual a 0 o menor que 0.
valor <- rnorm(1)
print (valor)
if (valor > 0) {
  print("Número mayor a 0") }else if (valor == 0) { print("Número igual a 0")
}else{
  print("Número menor a 0")
}
```

```r
# Este código asigna un valor fijo a una variable y utiliza una estructura 'if-else if-else' para verificar si el número se encuentra dentro de rangos específicos (0-2, 3-5) o fuera de ellos.
valor <- 4
print(valor)
if (valor >= 0 & valor <=2 ){
  print("Número está entre 0 y 2")
}else if (valor >= 3 & valor <=5) {
  print("Número está entre 3 y 5")
}else{
  print("El número ingresado NO se encuentra en dicho rango")
}
```

```r
# Este código utiliza un bucle 'while' para generar números aleatorios con distribución normal y los imprime, continuando el bucle hasta que se genere un número mayor o igual a 1.
valor <- rnorm(1)
while(valor <1) {
  valor <-rnorm(1)
  print(valor)
}
```

```r
# Este código utiliza un bucle 'repeat' para generar números aleatorios con distribución normal y los imprime, saliendo del bucle una vez que se genera un número menor que 1.
repeat {
  valor <- rnorm(1)
  print(valor)
  if (valor< 1) {
    break
  }
}
```

```r
# Este código inicializa vectores con nombres de jugadores y temporadas. Luego, crea dos matrices (`matrizGoles` y `matrizPartido`) a partir de los datos de goles y partidos, asignando los nombres de los jugadores como nombres de fila y las temporadas como nombres de columna. Finalmente, muestra la `matrizPartido`.
jugadores <- c("EstebanParedes", "LucasBarrios", "EduardoVargas","AlexisSanchez")
temporada <- c(2019, 2020, 2021, 2022)

#goles Jugadores
estebanGoles<-c(15,12,14,20)
lucasGoles <-c(7,9,12,4)
eduardoGoles <-c(2,4,5,6)
alexisGoles<-c(12,22,31,15)

#partidos Jugadores
estebanPartido <-c(20,22,23,20)
lucasPartido <-c(15,22,31,30)
eduardoPartido <- c(12,44,11,30)
alexisPartido <- c(33,12,30,19)

matrizGoles <- rbind(estebanGoles, lucasGoles, eduardoGoles, alexisGoles)
rownames (matrizGoles) <-c(jugadores)
colnames (matrizGoles)<-c(temporada)

matrizPartido <- rbind(estebanPartido, lucasPartido, eduardoPartido, alexisPartido)
rownames (matrizPartido)<-c(jugadores)
colnames (matrizPartido)<-c(temporada)
matrizPartido
```

```r
# Este código calcula la razón de partidos por gol para cada jugador y temporada, dividiendo `matrizPartido` por `matrizGoles` elemento a elemento, y redondea los resultados a un decimal.
round(matrizPartido/matrizGoles,1)
```

```r
# Este código muestra la `matrizGoles` original y luego su transpuesta (`t(matrizGoles)`), intercambiando filas por columnas.
matrizGoles
t(matrizGoles)
```

```r
# Este código genera un gráfico de líneas (`matplot`) a partir de la transpuesta de `matrizGoles`, usando diferentes tipos de puntos y colores. También añade una leyenda en la parte superior izquierda para identificar a cada jugador.
matplot(t(matrizGoles),type = "b",pch=15:20,col=c(1:4,6))
legend("topleft",inset =0.01, legend = jugadores, pch = 15:20, col=c(1:4,6), horiz = F)
```

```r
# Este código demuestra cómo seleccionar subconjuntos de la `matrizGoles`: las primeras dos filas, las primeras dos filas con las columnas 3 y 4, y filas y columnas no contiguas (filas 1 y 3, columnas 2 y 4).
matrizGoles[1:2,]
matrizGoles[1:2,3:4]
matrizGoles[c(1,3),c(2,4)]
```

```r
# Este código selecciona un subconjunto específico de `matrizGoles` (filas 1 y 3, columnas 2 y 4), lo guarda en `valores`, y luego genera un `matplot` con una leyenda para visualizar estos datos.
valores <- matrizGoles[c(1,3),c(2,4)]
valores
matplot(t(valores),type="b",pch=15:20,col=c(1:4,6))
legend("topleft", inset = 0.01, legend = jugadores[1:2],pch= 15:20,col=c(1:4,6), horiz = F)
```

```r
# Este código define una función llamada `generarGrafico` que toma datos de matriz, una leyenda, y especificaciones de filas y columnas para crear un `matplot` con leyenda. Luego, llama a esta función para generar un gráfico específico utilizando `matrizPartido`.
generarGrafico <- function(datos, legenda, filas, columnas) {
  valores <-datos [c(filas), c(columnas)]
  matplot (t(valores), type="b",pch=15:20, col=c(1:4,6))
  legend("topleft", inset= 0.01, legend = legenda [filas], pch=15:20, col=c(1:4,6), horiz = F)
}
generarGrafico(matrizPartido, jugadores, 2:3, 3:4)
```

```r
# Este código carga la librería `ggplot2`, lee el archivo 'titanic.csv' en `valoresTitanic`, muestra las primeras 5 columnas del dataframe, y luego crea un gráfico de barras básico (`geom_bar`) mostrando la distribución de la columna 'Pclass'.
library(ggplot2)
valoresTitanic <- read.csv('sample_data/titanic.csv')
head(valoresTitanic [,c(1:5)])
ggplot(data=valoresTitanic, aes(x = Pclass))+geom_bar()
```

```r
# Este código utiliza `ggplot2` para generar un gráfico de barras de la columna 'Pclass', donde cada barra está coloreada según el factor de 'Pclass' mismo, lo que permite visualizar la distribución de cada clase.
ggplot(data = valoresTitanic, aes(x=Pclass,fill=as.factor(Pclass)))+geom_bar()
```

```r
# Este código genera un gráfico de barras mostrando la distribución de 'Pclass', con las barras divididas y coloreadas por 'Sex'. Además, añade etiquetas claras para el eje Y ("Cantidad Pasajeros"), el eje X ("Clase turística"), la leyenda ("Sexo") y un título ("Supervivencia en Titanic por Clase").
ggplot (valoresTitanic, aes (x = Pclass, fill= as.factor(Sex))) + geom_bar() +
labs (y = "Cantidad Pasajeros", x = "Clase turística", fill="Sexo",title = "Supervivencia en Titanic por Clase")
```

```r
# Este código crea un gráfico de barras similar al anterior, mostrando la distribución de 'Pclass' por 'Sex', pero aplica el tema clásico `theme_classic()` para un estilo visual diferente. También incluye etiquetas personalizadas y un título.
ggplot(valoresTitanic, aes (x = Pclass, fill= as.factor(Sex))) + geom_bar() + theme_classic() +
labs (y = "Cantidad Pasajeros", x = "Clase turística", fill="Sexo",title = "Supervivencia en Titanic por Clase")
```

```r
# Este código genera un gráfico de barras facetado que muestra la relación entre 'Sex' y 'Survived' para cada 'Pclass'. Utiliza `facet_wrap(~ Pclass)` para crear un panel separado por cada clase, aplica el tema `theme_linedraw()` y añade etiquetas personalizadas y un título.
ggplot(valoresTitanic, aes (x = Sex, fill= as.factor(Survived))) + facet_wrap(~ Pclass)+ geom_bar() + theme_linedraw() +
labs (y = "Cantidad Pasajeros", x = "Clase turística", fill="Supervivencia",title = "Supervivencia en Titanic por Clase x Sexo")
```

```r
# Este código crea un gráfico de densidad facetado para visualizar la distribución de 'Age' (Edad), diferenciada por 'Survived' (Supervivencia) y organizada por 'Sex' y 'Pclass'. Utiliza `facet_wrap(Sex ~ Pclass)` para los paneles, aplica `geom_density()`, `theme_linedraw()` y proporciona etiquetas y un título descriptivos.
ggplot(valoresTitanic, aes(x = Age, fill= as.factor(Survived))) + facet_wrap(Sex ~ Pclass)+ geom_density() + theme_linedraw() +
labs (y = "Edad", x = "Supervivencia", fill="Supervivencia",title = "Supervivencia por Edad, Clase y Sexo")
```
